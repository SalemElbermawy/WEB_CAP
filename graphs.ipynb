```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
from matplotlib.dates import AutoDateLocator, ConciseDateFormatter

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(13, 9), sharex=True)
fig.patch.set_facecolor("#F8F8F6")

sampled = fixed_data[::800]

# ── Subplot 1: Temperature & Distance ────────────────────────────
ax1_r = ax1.twinx()

ax1.plot(sampled["Timestamp"], sampled["Temp_C"],
    color="#C0392B", linewidth=1.8, marker="o", markersize=4.5,
    markerfacecolor="white", markeredgewidth=1.4, label="Temp_C (°C)", zorder=3)

ax1_r.plot(sampled["Timestamp"], sampled["Distance_CM"],
    color="#1A6B3C", linewidth=1.8, marker="s", markersize=4.5,
    markerfacecolor="white", markeredgewidth=1.4,
    linestyle="--", label="Distance_CM (cm)", zorder=3)

ax1.set_ylabel("Temperature (°C)", fontsize=11, color="#C0392B", labelpad=8)
ax1_r.set_ylabel("Distance (cm)",  fontsize=11, color="#1A6B3C", labelpad=8)
ax1.tick_params(axis="y", colors="#C0392B", labelsize=10)
ax1_r.tick_params(axis="y", colors="#1A6B3C", labelsize=10)
ax1.yaxis.set_major_formatter(mticker.FormatStrFormatter("%.1f"))
ax1_r.yaxis.set_major_formatter(mticker.FormatStrFormatter("%d"))

for ax in [ax1, ax1_r]:
    ax.set_facecolor("#F8F8F6")
    ax.spines["top"].set_visible(False)
    ax.grid(False)

ax1.grid(True, linestyle="--", linewidth=0.5, color="#CCCCBB", alpha=0.9)
ax1.spines["left"].set_color("#C0392B")
ax1.spines["right"].set_visible(False)
ax1_r.spines["right"].set_color("#1A6B3C")
ax1_r.spines["top"].set_visible(False)
ax1_r.spines["left"].set_visible(False)

lines1, lab1 = ax1.get_legend_handles_labels()
lines2, lab2 = ax1_r.get_legend_handles_labels()
ax1.legend(lines1 + lines2, lab1 + lab2, fontsize=10,
    frameon=True, framealpha=0.95, edgecolor="#CCCCBB", loc="upper right")
ax1.set_title("Temperature & Distance over Time", fontsize=12, pad=10, color="#333333")

# ── Subplot 2: Lux (own scale, much larger range) ─────────────────
ax2.plot(sampled["Timestamp"], sampled["Lux"],
    color="#1A56A0", linewidth=1.8, marker="o", markersize=4.5,
    markerfacecolor="white", markeredgewidth=1.4, label="Lux (lx)", zorder=3)

ax2.set_ylabel("Illuminance (lx)", fontsize=11, color="#1A56A0", labelpad=8)
ax2.set_xlabel("Time", fontsize=11, labelpad=8)
ax2.tick_params(axis="y", colors="#1A56A0", labelsize=10)
ax2.tick_params(axis="x", labelsize=10, rotation=45)
ax2.yaxis.set_major_formatter(mticker.FuncFormatter(
    lambda x, _: f"{x:,.0f}"))          # e.g. 12,500

ax2.set_facecolor("#F8F8F6")
ax2.grid(True, linestyle="--", linewidth=0.5, color="#CCCCBB", alpha=0.9)
ax2.spines["top"].set_visible(False)
ax2.spines["right"].set_visible(False)
ax2.spines["left"].set_color("#1A56A0")

ax2.legend(fontsize=10, frameon=True, framealpha=0.95,
    edgecolor="#CCCCBB", loc="upper left")
ax2.set_title("Lux over Time", fontsize=12, pad=10, color="#333333")

# ── Shared x-axis formatting ──────────────────────────────────────
locator   = AutoDateLocator()
formatter = ConciseDateFormatter(locator)
ax2.xaxis.set_major_locator(locator)
ax2.xaxis.set_major_formatter(formatter)

fig.suptitle("Sensor Readings over Time", fontsize=14,
    fontweight="medium", color="#222222", y=1.01)

plt.tight_layout()
plt.show()
```
