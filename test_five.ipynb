```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
data=pd.read_csv("data_sets\whole_five.csv")
data.head()
```

```python
x_axis=np.arange(1,16)
x_axis
```

```python

sns.lineplot(x=x_axis, y=data["Lux"], label="Lux", linewidth=2,marker="o",color="green")

plt.xlabel("Time", fontsize=12)
plt.ylabel("Values", fontsize=12)

plt.xticks(rotation=45)
plt.yticks(fontsize=10)

plt.legend(title="Features", fontsize=10)

plt.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

```python
sns.lineplot(x=x_axis, y=data["Temp_C"], label="Temp_C", linewidth=2,marker="o",color="green")

sns.lineplot(x=x_axis, y=data["Distance_CM"], label="Distance_CM", linewidth=2,marker="o",color="green")


plt.xlabel("Time", fontsize=12)
plt.ylabel("Values", fontsize=12)

plt.xticks(rotation=45)
plt.yticks(fontsize=10)

plt.legend(title="Features", fontsize=10)

plt.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```
