```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv("deviation_temperature.csv")




sns.lineplot(x=df["Start_Value_C"], y=df["Duration_sec"], label="Temp_Deviation", linewidth=2,marker="o",color="blue")



plt.xlabel("Start_Value_C", fontsize=12)
plt.ylabel("Duration_sec", fontsize=12)

plt.xticks(rotation=45)
plt.yticks(fontsize=10)

plt.legend(title="Features", fontsize=10)

plt.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```
